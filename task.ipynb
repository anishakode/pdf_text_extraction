```python
import fitz
import pandas as pd

variables = ['Name','Phone','Mobile','Email','Office Address','Residence Address']
out_df = pd.DataFrame(columns=variables)

def text_analysis(text, page):
    global out_df
    page_to_text = page.get_text("words")
    df = pd.DataFrame(data=page_to_text)
    name_row = df[df[4] == text]
    input_df = pd.DataFrame(columns=variables)
    if text == 'CA.':
        round = 4
    else:
        round = 2
    for itr in range(round):

        input = dict.fromkeys(variables, None)
        ind = name_row.index[itr]

        # Name
        name = df[df[5] == df.iloc[ind][5]]
        s = name[4].tolist()
        name_str = ' '.join(s)
        input["Name"] = name_str

        no_row = df[(df[4] == '[+91]') & (df.index > ind)]
        i = no_row.iloc[0][5]

        # Mobile and Phone 
        x = 0
        for x in range(3):
            mobile = df[(df[5] == i) & (df[6] == x)]
            s = mobile[4].tolist()
            if len(s) != 0:
                mobile = ' '.join(s)
                break
        phone = df[(df[5] == i) & (df[6] == x+1)]
        s = phone[4].tolist()
        phone = ' '.join(s)
        input["Mobile"] = mobile
        input["Phone"] = phone

        # Email
        email = df[(df[5] == i+1) & (df[6] == 0)]
        s = email[4].tolist()
        email = ' '.join(s)
        input["Email"] = email

        # Office Address
        off_adrss = df[(df[5] == i+2)]
        s = off_adrss[4].tolist()
        off_adrss = ' '.join(s)
        input["Office Address"] = off_adrss

        # Residence Address
        res_adrss = df[(df[5] == i+3)]
        s = res_adrss[4].tolist()
        res_adrss = ' '.join(s)
        input["Residence Address"] = res_adrss

        input_df = pd.DataFrame(input, index=[0])
        out_df = pd.concat([out_df, input_df], ignore_index = True)
        
doc = fitz.open(r'File.pdf')
num_pages = doc.page_count
for i in range(1,num_pages,2):
    page = doc.load_page(i)
    text_analysis('CA.', page)
page = doc.load_page(num_pages-1)
text_analysis('Shri', page)
out_df.to_csv('output.csv', index=False)
```
